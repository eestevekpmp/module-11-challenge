```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

```python
# Scrape the website
html = browser.html

# Create a Beautiful Soup Object
weather_soup = soup(html, 'html.parser')
weather_soup
```

```python
# Extract all rows of data
all_rows = weather_soup.find('table', class_='table')
all_rows
```

```python
# Extract data items from rows.
td = all_rows.find_all('td')

td
```

```python
# Extract 'td' items from rows.
for values in td:
    print(values.text)
```

```python
# Create an empty list.
all_rows = []
columns = []

# Loop through the headings.
headings = weather_soup.find_all('th')
for heading in headings:
    columns.append(heading.get_text())
    
#Loop through the scraped data to create a list of dictionaries.
rows = weather_soup.find_all('tr',class_='data-row')

for row in rows:
    data = []
    row_data = row.find_all('td')
    for output in row_data:
        data.append(output.text)
    all_rows.append(data)
    
print(all_rows)
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names.
mars_weather_DF = pd.DataFrame(all_rows, columns = columns)
```

```python
# Confirm DataFrame was created successfully
mars_weather_DF.head()
```

```python
# Examine data type of each column
mars_weather_DF.dtypes
```

```python
# Change data types for data analysis

mars_weather_DF[mars_weather_DF.columns[1]] = pd.to_datetime(mars_weather_DF[mars_weather_DF.columns[1]])
mars_weather_DF[mars_weather_DF.columns[2]] = mars_weather_DF[mars_weather_DF.columns[2]].astype(int)
mars_weather_DF[mars_weather_DF.columns[3]] = mars_weather_DF[mars_weather_DF.columns[3]].astype(int)
mars_weather_DF[mars_weather_DF.columns[4]] = mars_weather_DF[mars_weather_DF.columns[4]].astype(int)
mars_weather_DF[mars_weather_DF.columns[5]] = mars_weather_DF[mars_weather_DF.columns[5]].astype(float)
mars_weather_DF[mars_weather_DF.columns[6]] = mars_weather_DF[mars_weather_DF.columns[6]].astype(float)
```

```python
# Confirm type changes were successful by examining data types again
mars_weather_DF.dtypes
```

```python
# 1. How many months are there on Mars?

print(f"There are {mars_weather_DF['month'].max()} months on Mars.")
```

```python
# 2. How many Martian days' worth of data are there?

print(f"There are {mars_weather_DF['sol'].max()} days' worth of data on Mars.")
```

```python
# 3. What is the average low temperature by month?

avg_low_temp_month = mars_weather_DF[['month','min_temp']].groupby(by='month').mean()
avg_low_temp_month
```

```python
# Plot the average temperature by month
fig, ax = plt.subplots(figsize=[9,6])
ax = plt.bar(avg_low_temp_month.index,avg_low_temp_month['min_temp'], color = 'blue')
plt.xlabel('Month')
plt.ylabel('Temperature (Degrees Celsius)')
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location
print(f"Based on Curiosity's location, the coldest and hottest months are {avg_low_temp_month['min_temp'].idxmin()} and {avg_low_temp_month['min_temp'].idxmax()} respectively.")
```

```python
# 4. Average pressure by Martian month
avg_press_month = mars_weather_DF[['month','pressure']].groupby(by='month').mean()
avg_press_month
```

```python
# Plot the average pressure by month
fig, ax = plt.subplots(figsize=[9,6])
ax = plt.bar(avg_press_month.index,avg_press_month['pressure'], color = 'orange')
plt.xlabel('Month')
plt.ylabel('Atmospheric Pressure')
plt.show()
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?
fig, ax = plt.subplots(figsize=[12,7])
ax = plt.plot(mars_weather_DF['sol'],mars_weather_DF['min_temp'])
plt.xlabel('Number of Terrestrial Days')
plt.ylabel('Minimum Temperature (Degrees Celsius)')
plt.title('Daily Minimum Temperature')
plt.show()
```

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.

```python
# Write the data to a CSV
mars_weather_DF.to_csv('mars_weather_DF.csv', index=False)
```
